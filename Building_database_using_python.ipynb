psycopg2 is a package which connects Python with PostgreSQL

```python
import psycopg2 as pg
import pandas as pd
```

Defining a function to Create Database, as we have to build 3 tables, it is better to make a function for that, rather

```python
def create_database():
    # connecting with default database in postgres, every databse have this kind info before login, it helpus to connect with them
    connection = pg.connect("host = localhost dbname = postgres user=postgres password=s1234")
    # A database cursor acts like a pointer that allows programmers to process and move through the results of a query one row at a time.
    cursor = connection.cursor()
    # We have to commit the changes we have made in our database so, we can make it automatic when we make some changes
    connection.set_session(autocommit = True)
    
    #lets create a database with UTF8 encoding
    cursor.execute("DROP DATABASE countries_db") # it is a good idea to delete any db named as Contries before creating one
    cursor.execute("CREATE DATABASE countries_db")

    # close connection to the default database because we will connect to the Countries db now
    connection.close()

    # connecting to the Countries Database
    connection = pg.connect("host = localhost dbname = countries_db user=postgres password=s1234")
    cursor = connection.cursor()
    return cursor, connection
```

Lets see what data we are going to add in our first table. we are not going to add all the attributes and tuples.

Now lets create the countries_db database. in which we will have 3 tables : countries, states, cities

```python
cursor, connection = create_database()
```

So basically what happens is we make the SQL Query inside of a variable and then passing the variable to cursor.execute(), so that the sql query get executed. 
- For being more sure, though we already setted the autocommit = True, lets do this again.
- Like that, lets add remaining 2 table like this.
- Then we will loop through the each corresponding dataframe to make it inside the table.
- This looping through is not the original process, how we enter data inside the database/datawarehouse. We will get to know that later.

```python
countries_table_create = (""" CREATE TABLE IF NOT EXISTS countries(
id INT PRIMARY KEY,
name VARCHAR ,
iso2 VARCHAR,
phone_code VARCHAR,
capital VARCHAR,
currency_name VARCHAR,
region VARCHAR,
nationality VARCHAR)""")
```

```python
cursor.execute(countries_table_create)
connection.commit()
```

```python
states_table_create = ("""CREATE TABLE IF NOT EXISTS states(
id int PRIMARY KEY,
name VARCHAR,
country_id INT,
country_code VARCHAR,
country_name VARCHAR,
state_code VARCHAR)""")
```

```python
cursor.execute(states_table_create)
connection.commit()
```

```python
cities_table_create = ("""CREATE TABLE IF NOT EXISTS cities(
id INT PRIMARY KEY,
name VARCHAR,
state_id INT,
state_code VARCHAR,
country_id INT,
country_code VARCHAR)""")
```

```python
cursor.execute(cities_table_create)
connection.commit()
```

Now, we have our table ready. We only need to enter data into it.

```python
countries_table_insert = ("""INSERT INTO countries(
id,
name,
iso2,
phone_code,
capital,
currency_name,
region,
nationality)
VALUES (%s,%s,%s,%s,%s,%s,%s,%s)""")
```

```python
for i, row in cleaned_countries_df.iterrows():
    cursor.execute(countries_table_insert, list(row))
```

```python
states_table_insert = ("""INSERT INTO states(
id ,
name ,
country_id ,
country_code ,
country_name ,
state_code )
VALUES (%s,%s,%s,%s,%s,%s)""")
```

```python
for i, row in cleaned_states_df.iterrows():
    cursor.execute(states_table_insert, list(row))
```

```python
cities_table_insert = ("""INSERT INTO cities(
id ,
name ,
state_id ,
state_code ,
country_id ,
country_code )
VALUES (%s,%s,%s,%s,%s,%s)""")
```

```python
for i, row in cleaned_cities_df.iterrows():
    cursor.execute(cities_table_insert, list(row))
```
